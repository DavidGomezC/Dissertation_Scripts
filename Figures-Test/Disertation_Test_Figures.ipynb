```python
l = (list(range(16)))

r = []

for x in l:
    x = Test = pd.read_csv('tests'+str(x+1)+'.csv', delimiter=",")
    r.append(x)
```

```python
A = []
B = []

for x in l:
    a = []
    b = []
    for row in r[x]:
        a.append(row)
        
    for y in a:
        b.append(r[x].at[0,y])
    A.append(a)
    B.append(b)

```

```python
plt.figure(dpi = 600)
for x in l:
    plt.plot(A[x],B[x],linestyle='steps',label='Trial '+str(x+1))
    plt.title('Convergence Curves')
    plt.yscale('symlog')
    plt.legend(bbox_to_anchor=(1.05, 1))
    plt.xlabel('Function Evaluations')
    plt.ylabel('Cost Function')
    plt.savefig("Convergence_Curve_Test")
plt.show()
    
```

```python
InputP = pd.read_csv('InputP.csv', delimiter=",")
InputSing = pd.read_csv('InputSing.csv', delimiter=",")

i = []
v = list(range(15))
w = list(range(3000))
t = list(range(100,3000,200))
val = []

for row in w:
    i.append(InputP.at[row, '0'])

for q in v:
    val.append(InputSing.at[q, '0'])
    
plt.figure(figsize=(6,5),dpi = 200)
plt.plot(i)
plt.title('OID')
plt.xlabel('Time (s)')
plt.ylabel('[IPTG]')
for i, q in enumerate(t):
    plt.scatter(q,1, label = val[i])
plt.legend(bbox_to_anchor=(1, 1))
plt.savefig("InputProfile_Test")
plt.show()
```
