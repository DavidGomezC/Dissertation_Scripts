FITNESS DEGRADATION

```python
FD = pd.read_csv('TestFitnessDegradation.csv', delimiter=",")
```

```python
fd1 = []
rl1 = []
fd2 = []
rl2 = []
v = list(range(len(FD)))

for row in v:
    fd1.append(FD.at[row, '1'])
    rl1.append(FD.at[row, '2'])
    fd2.append(FD.at[row, '3'])
    rl2.append(FD.at[row, '4'])
```

```python
cd Fitness_Degradation
```

```python
plt.figure(dpi = 400)

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

ax1.plot(v,fd1, label = 'Accuracy 1')
ax1.plot(v, rl1, label = 'Regression Line 1')
ax1.set_title('Beta Degradation 1: 2.7139e-07')
ax1.set_ylabel('Accuracy')
ax1.legend()

ax2.plot(v,fd2, label = 'Accuracy 2')
ax2.plot(v, rl2, label = 'Regression Line 2')
ax2.set_title('Beta Degradation 2: -2.537267e+29')
ax2.set_xlabel('Function Evaluations')
ax2.set_ylabel('Accuracy')
ax2.legend()

plt.show()
```
